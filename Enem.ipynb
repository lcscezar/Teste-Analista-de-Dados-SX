```python
### Bibliotecas
import pandas as pd
```

```python
### Extração de dados
arquivo = "/content/MICRODADOS_ENEM_2020.csv"
enem = pd.read_csv(arquivo, encoding="ISO-8859-1", sep = ';')
```

```python
#Filtrando as colunas

enem.columns.values
```

```python
### Selecionando as colunas necessárias

DadosEnem = ['NU_INSCRICAO', 'TP_FAIXA_ETARIA', 'TP_SEXO',
       'TP_COR_RACA',
       'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO',
       'TP_ESTADO_CIVIL', 'IN_TREINEIRO', 'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC',
       'CO_UF_ESC', 'SG_UF_ESC', 'TP_DEPENDENCIA_ADM_ESC',
       'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC', 'CO_MUNICIPIO_PROVA',
       'NO_MUNICIPIO_PROVA', 'CO_UF_PROVA', 'SG_UF_PROVA',
       'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC',
       'TP_PRESENCA_MT', 'NU_NOTA_REDACAO','NU_NOTA_MT',
    'NU_NOTA_CN',
    'NU_NOTA_CH',
    'NU_NOTA_LC',
    'NU_NOTA_REDACAO']
```

```python
### Atribuindo as legendas aos campos

### Cor/Raça
desc_raca = {
               0 : "Não Declarado",
               1 : "Branco",
               2 : "Preto",
               3:  "Pardo",
               4 : "Amarelo",
               5 : "Indígena"
               }

enem["TP_COR_RACA"] = enem["TP_COR_RACA"].replace(desc_raca)

### Estado Civil
desc_estado_civil = {
               0 : "Não Informado",
               1 : "Solteiro(a)",
               2 : "Casado(a)/Mora com companheiro(a)",
               3:  "Divorciado(a)/Desquitado(a)/Separado(a)",
               4 : "Viúvo(a)"
               }

enem["TP_ESTADO_CIVIL"] = enem["TP_ESTADO_CIVIL"].replace(desc_estado_civil)


### Situação Conclusão
desc_conclusao = {
               1 : "Já concluí o Ensino Médio",
               2 : "Estou cursando e concluirei o Ensino Médio em 2020",
               3 : "Estou cursando e concluirei o Ensino Médio após 2020",
               4 : "Não concluí e não estou cursando o Ensino Médio"
               }

enem["TP_ST_CONCLUSAO"] = enem["TP_ST_CONCLUSAO"].replace(desc_conclusao)


### Tipo de escola do Ensino Médio
desc_escola = {
               1 : "Não Respondeu",
               2 : "Pública",
               3 : "Privada",
               4:  "Exterior"
               }

enem["TP_ESCOLA"] = enem["TP_ESCOLA"].replace(desc_escola)


### Tipo de instituição que concluiu ou concluirá o Ensino Médio
desc_ensino = {
               1 : "Ensino Regular",
               2 : "Educação Especial - Modalidade Substitutiva",
               3:  "Educação de Jovens e Adultos"
               }

enem["TP_ENSINO"] = enem["TP_ENSINO"].replace(desc_ensino)

### Indica se o inscrito fez a prova com intuito de apenas treinar seus conhecimentos
desc_treineiro = {
               1 : "Sim",
               0 : "Não"
               }

enem["IN_TREINEIRO"] = enem["IN_TREINEIRO"].replace(desc_treineiro)

### Dependência administrativa (Escola)
desc_dependencia = {
               1 : "Federal",
               2 : "Estadual",
               3 : "Municipal",
               4 : "Privada"
               }

enem["TP_DEPENDENCIA_ADM_ESC"] = enem["TP_DEPENDENCIA_ADM_ESC"].replace(desc_dependencia)

### Localização (Escola)
desc_localizacao = {
               1 : "Urbana",
               2 : "Rural"
               }

enem["TP_LOCALIZACAO_ESC"] = enem["TP_LOCALIZACAO_ESC"].replace(desc_localizacao)

### Situação de funcionamento (Escola)
desc_situacao = {
               1 : "Em atividade",
               2 : "Paralisada",
               3 : "Extinta",
               4 : "Escola extinta em anos anteriores"
               }

enem["TP_SIT_FUNC_ESC"] = enem["TP_SIT_FUNC_ESC"].replace(desc_situacao)

print(desc_raca)
print(desc_estado_civil)
print(desc_conclusao)
print(desc_escola)
print(desc_ensino)
print(desc_treineiro)
print(desc_dependencia)
print(desc_localizacao)
print(desc_situacao)
```

```python
### Separando as colunas necessárias

EnemDados = enem.filter(DadosEnem)

EnemDados.head(10)
```

```python
### Removendo os valores ausentes da base

EnemDadosIn = EnemDados.dropna()
```
